```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
df.head()
```

```python
#Q1
df1=df['total_profit']
df_m=df['month_number']
plt.plot(df_m,df1,marker='o')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title("Month-Profit Graph")
```

```python
#Q2
plt.plot(df_m,df1,marker='o',linestyle='dotted',color='r',linewidth=3)
lab=['Units sold']
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title("Month-Profit Graph")
plt.legend(loc='lower right',labels=lab)
plt.tight_layout()
```

```python
#q3
for i in df.columns[1:-2]:
    plt.plot(df["month_number"],df[i],marker='s',mfc='g')
plt.xlabel('Month Number')
plt.ylabel('Units sold')
plt.show()
```

```python
#Q4
plt.scatter(df['month_number'],df['toothpaste'])
plt.xlabel('Month Number')
plt.ylabel('Units sold')
```

```python

#Q5
plt.bar(df['month_number'],df['facecream'])
plt.bar(df['month_number'],df['facewash'])
lab=['Facecream','Facewash']
plt.legend(labels=lab)
plt.tight_layout()
plt.show()
```

```python
#q6
plt.hist(df['total_profit'])
```

```python
#Q7
x=df.drop(columns=['month_number','total_units','total_profit']).sum()
lab=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie(x,labels=lab,autopct='%1.0f%%')
```
