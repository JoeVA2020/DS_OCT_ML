```python
Team1_runs=np.array([9,5,12,5,6,8,4,20,1,9,10,7,8,15,2,9,6,7,21,1])
Team1_wicket=np.array([0,0,1,0,0,2,0,0,0,2,0,0,0,0,1,0,0,0,1,3])
Team2_runs = [16, 10, 12, 8, 10, 12, 10, 6, 8, 18, 20, 2, 8, 8, 14, 10, 8, 10, 12, 16]
Team2_wickets = [0, 0, 1, 0, 0, 0, 0, 0, 2, 0, 0, 0, 3, 0, 0, 0, 0, 0, 0, 1]
```

```python
plt.bar(range(1,len(Team1_runs)+1),Team1_runs,label='Runs',width=.5)
wicket_count=0
for i in range(len(Team1_wicket)):
    if Team1_wicket[i]>0:
        for j in range(1,Team1_wicket[i]+1):
            wicket_count+=1
            if wicket_count!=sum(Team1_wicket):
                plt.scatter(i+1,Team1_runs[i]+j, color='red')
            else:
                plt.scatter(i+1,Team1_runs[i]+j, color='red',label="Wickets")
                
                
plt.xticks(range(1,len(Team1_runs)+1))
plt.yticks(range(0,30,4))
plt.title('Score Analysis', fontsize=15, fontweight='bold', fontstyle='italic')
plt.xlabel('Overs')
plt.ylabel('Runs per over')
plt.legend()
plt.show()
```
