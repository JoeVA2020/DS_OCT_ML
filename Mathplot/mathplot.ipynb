```python
x1=np.array([1,2,3,4,5,6])
y1=np.array([3,8,1,10,5,7])
plt.plot(x1,y1,marker='o')
```

```python
plt.plot(x1,y1,'p--c')
```

```python
#ms : marker size
#mfc : marker face color
#mec : marker edge color
plt.plot(x1,y1,marker='o',ms=15,mfc='r',mec='g')
```

```python
# LINE PROPERTIES
# linestyle : dotted,dashed,solid(default)
# linewidth : width of line
plt.plot(x1,y1,marker='o',linestyle='dotted',linewidth=3,color='k')
```

```python
x1=[1,2,3,4]
y1=[3,7,1,9]
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'b')
```

```python
height=[167,172,176,180]
weight=[65,70,75,80]
plt.plot(height,weight,marker='o',mfc='b',ms=5,mec='b',color='r')
plt.xlabel('Height in cm')
plt.ylabel('weight in Kg')
plt.title('Height-Weight Graph')
```

```python
x=[50,70,80,100]
y=[50,68,91,120]
plt.plot(x,y,marker='o',ms=8,mfc='r',mec='r')
plt.xlabel('Velocity in m/s')
plt.ylabel('Displacement in m')
plt.title('Velocity - Displacement Graph')
plt.grid(color='g')
# plt.grid(axis='x') : shows gridline perpendicular to x axis
# plt.grid(axis='y') : shows gridline perpendicular to y axis
```

```python
# Scatterplot
x=[10,2,4,6,3,7,8,2,1]
y=[8,10,11,7,9,13,5,7,2]
x1=[5,6,7,1,2,3]
y1=[7,6,2,6,9,10]
plt.scatter(x,y,color='r',)
plt.scatter(x1,y1,color='b')
```

```python
# Subplot
# partioning a graph to accomidate another graph
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1) 
#SYNTAX : plt.subplot(row,column,position of graph)
plt.plot(x1,y1)

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)
plt.plot(x2,y2)
```

```python
# partioning a graph to accomidate another graph
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,2) 
#SYNTAX : plt.subplot(row,column,position of graph)
plt.plot(x1,y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph 1')

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,1)
plt.plot(x2,y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph 2')
plt.suptitle('Graph 1 - Graph 2')
```

```python
# partioning a graph to accomidate another graph
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(2,1,1) 
#SYNTAX : plt.subplot(row,column,position of graph)
plt.plot(x1,y1)

x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(2,1,2)
plt.plot(x2,y2)
```

```python
# Barchart
x1=[3,4,6,1]
y1=[1,7,2,8]
plt.bar(x1,y1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph 1')

# USING barh : gives horizontal bar_chart
```
