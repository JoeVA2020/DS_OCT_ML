```python
df['Seats'].value_counts()
```

```python
df['Seats']=df['Seats'].fillna(df['Seats'].mode()[0])
```

```python
df['Engine'].unique()
```

```python
df['Engine']=df['Engine'].str.replace('CC','0')
df['Engine'].unique()
```

```python
df['Mileage']=df['Mileage'].apply(lambda x:str(x).replace('kmpl','0'))
df['Mileage']=df['Mileage'].apply(lambda x:str(x).replace('km/kg','0'))
df['Power']=df['Power'].apply(lambda x:str(x).replace('bhp','0'))
df
```

```python
df['Brands']=df['Name'].apply(lambda x:x.split(' ')[0])
df['Brands'].unique()
```

```python
df.drop(columns=['Unnamed: 0','Name','New_Price'],inplace=True)
df
```

```python
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brands']],drop_first=True,dtype=int)
df1
```

```python
dfe=pd.concat((df1,df.drop(columns=['Location','Fuel_Type','Transmission','Owner_Type','Brands'],axis=1)),axis=1)
dfe
```

```python
dfe['Power'].unique()
```

```python
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe.isna().sum()
```
