```python
df=pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel('Years od Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience Graph')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[7.9]]))
```

```python
plt.scatter(x_train,y_train,label='training datapoints')
plt.plot(x_test,y_pred,color='g',label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience Graph')
plt.legend()
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'difference':y_test-y_pred})
df1
```

```python
print("Slope is",model.coef_)
```

```python
print('constant is',model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
RMSE=root_mean_squared_error(y_test,y_pred)
RMSE
```

```python
score=r2_score(y_test,y_pred)
score
```
