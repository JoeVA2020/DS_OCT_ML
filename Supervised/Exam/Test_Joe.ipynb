```python
# Dropping Cabin column due to missing values
df.drop("Cabin", axis=1, inplace=True)
```

```python
df['Pclass'].unique()
```

```python
df['Embarked'].unique()
```

```python
# filiing missing values in Age and Pclass
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Pclass"]=df["Pclass"].fillna(df["Pclass"].mode()[0])
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
df.info()
```

```python
Corrilation=df.corr(numeric_only=True)
sns.heatmap(Corrilation, annot=True, cmap="coolwarm")
```

```python
df["Ticket"].value_counts()
```

```python
# Dropping Unwanted columns 
df.drop(columns=["PassengerId","Name","Ticket"],axis=1,inplace=True)
df
```

```python
# encoding Sex and Embarked columns
cols=['Sex','Embarked']
dfe=pd.get_dummies(df[cols],dtype=int)
dfe
```

```python
dfe=pd.concat([df.drop(columns=["Sex","Embarked"]),dfe],axis=1)
dfe
```

```python
# seperate input and output data

x=dfe.drop("Survived",axis=1)
y=dfe["Survived"]
x
```

```python
# Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Scaing Data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=11)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print("Accuracy :",accuracy_score(y_test,y_pred))   
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred))
print("Classification Report :\n",classification_report(y_test,y_pred))
```

```python
# 1.	What is the survival rate for males vs. females?
plt.figure(figsize=(10, 5))
survival_rate = df.groupby('Sex')['Survived'].mean()
survival_rate.plot(kind='bar', color=['blue', 'pink'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()

# females have higher chance of survival
```

```python
#2.	Did passengers in higher classes (1st class) have a higher survival rate compared to lower classes (3rd class)?
plt.figure(figsize=(10, 5))
survival_rate = df.groupby('Pclass')['Survived'].mean()
survival_rate.plot(kind='bar', color=['Red','Blue','pink'])
plt.title('Survival Rate by Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks(rotation=0)
plt.show()
# yes. (1st class) have a higher survival rate compared to lower classes (3rd class)
```

```python
# 3.	Did passengers with more family members (siblings/spouses or parents/children) aboard have a higher or lower chance of survival?
plt.figure(figsize=(10, 5))
survival_rate = df.groupby('SibSp')['Survived'].mean()
survival_rate.plot(kind='bar', figsize=(8, 5))
plt.title('Survival Rate by Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.xlabel('Sibling / Spouse Count')
plt.xticks(rotation=0)
plt.show()

# Passengers with few or no silbligs/ Familymemebers had higher Survival
```

```python
#4.	Did the port of embarkation (C, Q, S) influence the survival rate?
labels=['Cherbourg','Queenstown','Southampton']
plt.figure(figsize=(10, 5))
survival_rate = df.groupby('Embarked')['Survived'].mean()
survival_rate.plot(kind='bar', color=['Red','Blue','pink'])
plt.title('Survival Rate by Embarked port')
plt.ylabel('Survival Rate')
plt.xlabel('Embarked Port')
plt.xticks(rotation=0)
plt.show()

# pansengers boarded from Cherbourg har higher survival
```
