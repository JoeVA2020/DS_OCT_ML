```python
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','brand']],drop_first=True,dtype=int)
df1
```

```python
df1.shape
```

```python
dfe=pd.concat((df,df1),axis=1)
dfe
```

```python
dfe.drop(columns=['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','brand','New_Price'],axis=0,inplace=True)
dfe
```

```python
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
dfe
```
