```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.compose import make_column_transformer #to convert into difference column
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'difference':y_test-y_pred})
df1
```

```python
print("slope is",list(zip(x,model.coef_)))
```

```python
print("constant is",model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
from sklearn.metrics import root_mean_squared_error
RMSE=root_mean_squared_error(y_test,y_pred)
RMSE
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
