Find percentage score of resume

```python
import cv2
import pytesseract
import re
```

```python
img = cv2.imread('JoeResumev1.jpg')
img
```

```python
cv2.imshow('Resume',img)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
resume_text = pytesseract.image_to_string(img)
print(resume_text)
```

```python
dic = pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
def preprocess_text(text):
    text = text.lower()  
    text = re.sub(r'\W+', ' ', text) # remove special characyers
    words = text.split()
    return words
```

```python
resume_words = preprocess_text(resume_text)
print(resume_words)
```

```python
skills = {"python", "numpy", "pandas" , "eda", "sql", "machine", "learning", "deep", "natural language processing", "data", "visualization", "DL", "ML"}
```

```python
matches = set(resume_words).intersection(skills)
print(matches)
```

```python
l = len(dic['text'])
for i in range(l):
    if dic['conf'][i] > 50:
        if dic['text'][i] in matches:
            x = dic['left'][i]
            y = dic['top'][i]
            w = dic['width'][i]
            h = dic['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('Resume',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
Total_Skills = len(skills)
match_percentage = (len(matches) / Total_Skills) * 100
print(f"Resume Percentage = {match_percentage:.2f}%")
```
