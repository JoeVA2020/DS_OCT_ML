```python
import cv2
import pytesseract
```

```python
img=cv2.imread('ocr3.jpg')
```

```python
cv2.imshow('IMG',img)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text=pytesseract.image_to_string(img)
text
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
data
```

```python
n=len(data['text'])
n
```

```python
count=0
for i in range(n): #0
    if  data['text'][i]=='dog':
        x=data['left'][i]
        y=data['top'][i]
        w=data['width'][i]
        h=data['height'][i]
        count+=1
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('img',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
