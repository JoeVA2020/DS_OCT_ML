```python
import cv2
import pytesseract
```

```python
img=cv2.imread('ocr1.jpg')
img
```

```python
cv2.imshow("Img1",img)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
# set path tesseract ocr

pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
# Extract data from image

text=pytesseract.image_to_string(img)
print(text)
```

```python
# Text properties
# output is formed in dictionary format

dic=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
dic
```

```python
x,y,w,h=43,75,210,32
cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('Image1',img)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
img2=cv2.imread('ocr3.jpg')
cv2.imshow('IMG2',img2)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
text2=pytesseract.image_to_string(img2)
print(text2)
```

```python
data=pytesseract.image_to_data(img2,output_type=pytesseract.Output.DICT)
data
```

```python
# to print the rectangle over the text we will use the key value pairs of the dictionary
n=len(data['text'])
n
```

```python
for i in range(n): #0
    if data['conf'][i]>50:
        x=data['left'][i]
        y=data['top'][i]
        w=data['width'][i]
        h=data['height'][i]
        cv2.rectangle(img2,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('img',img2)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
