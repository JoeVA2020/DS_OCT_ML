```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('alen.JPG')
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cas
```

```python
face=face_cas.detectMultiScale(conv_img,minNeighbors=15)
face
# No. of rows means the Number of faces detected
# Increasing minNeighbours increases the Number of times the image recognized as a face
# Value in the array can be interperted as [x1,y1,width,height]
```

```python
for(x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),4)
```
