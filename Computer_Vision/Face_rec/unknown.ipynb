```python
import numpy as np
import cv2
import os
import face_recognition as fr
import winsound
```

```python
path='images'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]

for i in mylist:
    imgpath=os.path.join(path,i)
    currentimg=cv2.imread(imgpath)
    imgs.append(currentimg)
    classnames.append(i.split('.')[0])
```

```python
classnames
```

```python
def faceencoding(images):
    encodelist=[]
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img)
        print(face_in_frame)
        encode=fr.face_encodings(img,face_in_frame)[0]
        encodelist.append(encode)
    return encodelist
```

```python
encodelist_knownfaces=faceencoding(imgs)
print(encodelist_knownfaces)
```

```python
video=cv2.VideoCapture(0)
while True:
    sucess,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(img1)
    #print(face_in_frame)
    
    encoded_face=fr.face_encodings(img1,face_in_frame)
    #print(encoded_face)
    
    for face_loc,enc in zip(face_in_frame,encoded_face):
        matches=fr.compare_faces(encodelist_knownfaces,enc)
        print(matches)
        face_dis=fr.face_distance(encodelist_knownfaces,enc)
        #print(face_dis)
        matchindex=np.argmin(face_dis)
        #print(matchindex)
        if matches[matchindex]==True:
            name=classnames[matchindex]
            #print(name)
        else:
            name='Unknown'
            winsound.Beep(1000,500)
            
        y1,x2,y2,x1=face_loc
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),2)
        cv2.putText(img,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)

            
                
    cv2.imshow("video",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
