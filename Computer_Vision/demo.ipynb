```python
# Convert Grey scale img to binary
# threshhold

thresh,bin_img=cv2.threshold(grey_image1,125,255,cv2.THRESH_BINARY)
cv2.imshow('Binary img',bin_img)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img=cv2.imread('k1.jpeg')
img
```

```python
plt.imshow(img)
```

```python
#BGR to RGB

conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
cv2.rectangle(conv_img,pt1=(210,45),pt2=(330,170),color=(255,0,0),thickness=3)
plt.imshow(conv_img)
```

```python
# rectangle corrected
cv2.rectangle(conv_img,pt1=(100,150),pt2=(330,250),color=(255,0,0),thickness=3)
plt.imshow(conv_img)
```

```python
conv_img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(conv_img1,pt1=(100,150),pt2=(330,250),color=(255,0,0),thickness=3)
plt.imshow(conv_img1)
```

```python
conv_img2a=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
conv_img2b=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img2a)
```

```python
# circle
# center
# radius
cv2.circle(conv_img2a,center=(300,200),radius=120,color=(0,255,0),thickness=3)
plt.imshow(conv_img2a)
```

```python
cv2.circle(conv_img2b,center=(275,300),radius=150,color=(255,255,0),thickness=5)
plt.imshow(conv_img2b)
```

```python
cv2.circle(conv_img2b,center=(275,300),radius=150,color=(255,255,0),thickness=-1)
plt.imshow(conv_img2b)
#can use negetive values in thickness to hide faces
```
