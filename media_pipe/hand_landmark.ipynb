```python
import mediapipe as mp
import cv2
```

```python
#Hand landmark detection
mphands=mp.solutions.hands #hands
mp_drawing=mp.solutions.drawing_utils 
```

```python
data=mphands.Hands()
```

```python
video=cv2.VideoCapture(1)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks :
        for x in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(img,x,mphands.HAND_CONNECTIONS,mp_drawing.DrawingSpec((0,255,0),thickness=5,circle_radius=4))
    
    cv2.imshow("Hands",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
