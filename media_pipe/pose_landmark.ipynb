```python
#pose landmark

import cv2
import mediapipe as mp
```

```python
mp_detect=mp.solutions.pose
mp_drawing=mp.solutions.drawing_utils
```

```python
pose_detect=mp_detect.Pose()
```

```python
video=cv2.VideoCapture(1)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=pose_detect.process(img1)
    print(result.pose_landmarks)
    if result.pose_landmarks:
        mp_drawing.draw_landmarks(img,result.pose_landmarks,mp_detect.POSE_CONNECTIONS)
            
    cv2.imshow("Face",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
